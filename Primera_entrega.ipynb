**GRAFICO DE BARRAS, TIPO DE ABUSOS VIOLENCIA SEXUAL**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargo el conjunto de datos
df_abuso = pd.read_csv('abuso.csv')

# Selecciono columnas relevantes para las formas de violencia sexual
columnas_vs = ['vs_violacion_via_vaginal', 'vs_violacion_via_anal', 'vs_violacion_via_oral',
               'vs_tentativa_violacion', 'vs_tocamiento_sexual', 'vs_intento_tocamiento',
               'vs_Intento_violación_tercera_persona', 'vs_violencia_digital', 'vs_exhibicionismo',
               'vs_amenazas_verbales_contenido_sexual', 'vs_explotacion_sexual',
               'vs_explotacion_sexual_comercial', 'vs_explotacion_sexual_viajes_turismo',
               'vs_sospecha_trata_personas_fines_sexuales', 'vs_existencia_facilitador_corrupcion_nnya',
               'vs_obligacion_sacarse_fotos_pornograficas', 'vs_eyaculacion_partes_cuerpo',
               'vs_acoso_sexual', 'vs_iniciacion_sexual_forzada_inducida', 'vs_otra_forma_violencia_sexual',
               'vs_no_sabe_no_contesta']

# Convierto los valores de las columnas a "SI" a 1 y "NO" a 0
df_abuso[columnas_vs] = df_abuso[columnas_vs].replace({'SI': 1, 'NO': 0})

# Calculo la frecuencia de cada forma de violencia sexual
frecuencia_vs = df_abuso[columnas_vs].sum()

# Y creo el gráfico de barras
plt.figure(figsize=(12, 6))
frecuencia_vs.plot(kind='bar', color='pink', edgecolor='black')
plt.title('Frecuencia de formas de violencia sexual')
plt.xlabel('Forma de violencia sexual')
plt.ylabel('Frecuencia')
plt.xticks(rotation=90)
plt.show()
```

**Diagrama de dispersión: Para ver la relación entre la edad de la víctima y del llamante.**

Primero, cargo el conjunto de datos y selecciono las columnas relevantes, que en este caso serían 'victima_edad' y 'llamante_edad'.
Luego, creo el diagrama de dispersión utilizando Matplotlib o Seaborn.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el conjunto de datos
df_abuso = pd.read_csv('abuso.csv')

# Seleccionar las columnas relevantes
edades_data = df_abuso[['victima_edad', 'llamante_edad']]

# Eliminar filas con valores faltantes
edades_data = edades_data.dropna()

# Crear el diagrama de dispersión
plt.figure(figsize=(30, 30))
sns.scatterplot(x='victima_edad', y='llamante_edad', data=edades_data, alpha=0.7)
plt.title('Relación entre la edad de la víctima y del llamante')
plt.xlabel('Edad de la víctima')
plt.ylabel('Edad del llamante')
plt.grid(True)
plt.show()
```

**GRAFICO DE LINEA DE TIEMPO:** Realizo un análisis temporal para identificar posibles patrones estacionales en la frecuencia de denuncias.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator
import calendar

# Convertir la columna 'llamado_fecha_hora' al tipo de datos datetime
df_abuso['llamado_fecha_hora'] = pd.to_datetime(df_abuso['llamado_fecha_hora'])

# Extraer el mes de la columna 'llamado_fecha_hora'
df_abuso['mes'] = df_abuso['llamado_fecha_hora'].dt.month

# Contar el número de denuncias por mes
denuncias_por_mes = df_abuso['mes'].value_counts().sort_index()

# Crear un índice de meses con nombres de meses en lugar de números
meses = [calendar.month_name[i] for i in range(1, 13)] #No se como hacer para que aparezcan los nombres de los meses
denuncias_por_mes.index = meses

# Graficar el número de denuncias por mes
plt.figure(figsize=(12, 6))
denuncias_por_mes.plot(kind='line', marker='o')
plt.title('Tendencia mensual de las denuncias de abuso sexual (2023)')
plt.xlabel('Mes')
plt.ylabel('Número de denuncias')

# Configurar el espaciado de los meses en el eje X
plt.gca().xaxis.set_major_locator(MonthLocator())
plt.xticks(rotation=90)  # Rotar las etiquetas de los meses para que sean legibles

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Suponiendo que 'df_abuso' es el nombre de tu DataFrame y 'victima_edad' es la columna que contiene la edad de la víctima

# Convertir la columna 'victima_edad' a tipo numérico
df_abuso['victima_edad'] = pd.to_numeric(df_abuso['victima_edad'], errors='coerce')

# Definir los rangos de edad
rangos_edad = [(0, 12), (12, 18), (19, 26), (27, 59), (60, float('inf'))]
etiquetas_edad = ['Niñez', 'Adolescencia', 'Juventud', 'Adultez', 'Persona Mayor']

# Crear una nueva columna que categorice la edad de la víctima
df_abuso['grupo_edad'] = pd.cut(df_abuso['victima_edad'], bins=[rango[0] for rango in rangos_edad] + [float('inf')], labels=etiquetas_edad)

# Contar la frecuencia de casos judicializados en cada grupo de edad
resumen_edad = df_abuso[df_abuso['caso_judicializado'] == 'SI']['grupo_edad'].value_counts()

print(resumen_edad)
```
